## Section 3 — Classify Logical Constraint Types

* Policy 1
  * IF–THEN rule: If user devices are patched, then critical servers must be patched.
  * Constraint type: Activation
  * Explanation: Selecting the user device patching decision requires the critical server patching decision to also be selected.
  * Clarification: No ambiguity—this is a direct dependency (A → B).
* Policy 2
  * IF–THEN rule: If critical servers are not patched, then MFA for admin accounts must be enabled.
  * Constraint type: Activation
  * Explanation: The absence of one control (patching) triggers the requirement to activate another control (MFA), which still fits an activation structure.
  * Clarification: Confirm whether MFA is required only when servers are unpatched or always required.
* Policy 3
  * IF–THEN rule: If a system is suspected to be compromised, then it must be isolated.
  * Constraint type: Activation
  * Explanation: Detection of compromise directly requires isolation, making this a mandatory follow-on action.
  * Clarification: Confirm whether “suspected” vs. “confirmed” compromise changes the requirement.
* Policy 4
  * IF–THEN rule: If firewall rules are updated, then network vulnerabilities must be remediated first.
  * Constraint type: Activation
  * Explanation: Updating firewall rules depends on prior remediation, meaning one decision requires another to be true.
  * Clarification: Clarify whether this is a strict requirement or a best practice.
* Policy 5
  * IF–THEN rule: If remediation actions are performed, then a security scan must be conducted.
  * Constraint type: Activation
  * Explanation: Any remediation decision triggers the requirement for a preceding scan, forming a dependency relationship.
  * Clarification: Confirm if the scan must occur before every action or once per cycle.
* Policy 6
  * IF–THEN rule: If all major remediation actions are selected, then at least one lower-priority action must not be selected.
  * Constraint type: Suppression
  * Explanation: Selecting all high-priority actions limits the ability to select additional lower-priority actions due to resource constraints.
  * Clarification: Confirm whether this should instead be modeled as a budget/resource constraint.
* Policy 7
  * IF–THEN rule: If firewall rules are updated, then MFA for admin accounts must be enabled.
  * Constraint type: Activation
  * Explanation: Updating firewall rules requires MFA to also be enabled, creating a direct dependency.
  * Clarification: Confirm whether MFA is strictly required or recommended.

## Section 5 — Feasibility, Conflicts, and "No Feasible Solution"

A realistic conflict appears between P1 (patch priority), P5 (scan before remediation), and P6 (resource constraint). P1 pushes the team to prioritize patching critical systems quickly, while P5 requires a security scan before any remediation can occur. At the same time, P6 limits how many actions can be performed. Together, these can create a situation where patching must happen immediately, but cannot proceed without a scan, and there are not enough resources to do both, resulting in no combination of decisions that satisfies all policies.

In practice, “no feasible solution” would look like a manager seeing urgent vulnerabilities but being unable to act because every option violates at least one rule. To resolve this, the most reasonable policy to relax would be P5, allowing critical patches to proceed without a full scan in high-risk situations. To a non-technical stakeholder, you could explain it as, "the rules are too restrictive to allow action, so one needs to be temporarily adjusted to move forward while still maintaining overall security priorities."
